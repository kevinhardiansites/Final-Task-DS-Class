```python
#Mengatur dataset yang akan digunakan
df= pd.read_csv(path_dataset)
```

```python
#Mengecek dataset setelah diatur
df.head()
```

```python
#Melihat tipe data dari tiap kolom
df.info()
```

```python
#Mengonversi tipe data baris 'reservation_status_date' menjadi datetime
df['reservation_status_date'] =pd.to_datetime(df['reservation_status_date'])
```

```python
df.info()
```

```python
#Menghitung jumlah nilai null di setiap kolom
df.isnull().sum()
```

```python
#Menghapus kolom-kolom yang memiliki nilai null
df.drop(['company','agent'], axis = 1, inplace=True)
df.dropna(inplace=True)
```

```python
##Menghitung jumlah nilai null di setiap kolom kembali
df.isnull().sum()
```

```python
df.describe()
```

```python
#Mencari outliers
df['adr'].plot(kind='box')
```

```python
#Membatasi data yang digunakan
df=df[df['adr']<5000]
```

```python
#Menampilkan grafik jumlah status reservasi
canceled_perc = df['is_canceled'].value_counts(normalize=True)
print(canceled_perc)

plt.figure(figsize=(5,4))
plt.title(
    'Jumlah Status Reservasi'
)
plt.bar(['Not canceled', 'canceled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.5)
plt.show()
```

```python
#Menampilkan grafik dari status reservasi hotel yang berbeda
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel', hue='is_canceled', data=df, palette ='Blues')
legend_labels,_= ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))

plt.title('status reservasi di tiap jenis hotel', size=20)
plt.xlabel('hotel')
plt.ylabel('jumlah reservasi')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#Mengambil rerata 'adr' dari tiap hotel
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
#Membuat grafik rerata 'adr' dalam plot
plt.figure(figsize =(20,8))
plt.title('Rerata pendapatan harian sewa kamar City Hotel and Resort Hotel', size=20)
plt.plot(resort_hotel.index, resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize =(16,8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette ='pastel')
legend_labels,_ =ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Status reservasi per bulan', size = 20)
plt.xlabel('month')
plt.ylabel('Jumlah reservasi')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
plt.figure(figsize =(15,8))
plt.title('Rerata pendapatan harian sewa kamar perbulan', fontsize=30)
sns.barplot(x='month', y='adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
canceled_data=df[df['is_canceled']==1]
top_10_country =canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 negara dengan pembatalan reservasi')
plt.pie(top_10_country, autopct='%.2f', labels=top_10_country.index)
```
